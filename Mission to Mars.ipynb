# NASA Mars News

Scrape the NASA Mars News Site and collect the latest News Title and Paragraph Text. Assign the text to variables that you can reference later.

```python
#dependencies
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import pymongo
import os
```

```python
#create connection for pymongo 
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
#news url
news_url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'
```

```python
#parse the HTML
response = requests.get(news_url)

soup = bs(response.text, 'html.parser')
#type(soup)
print(soup)
```

```python
#span HTML for first article 
first_article = soup.find('title').text
    #find('a', href="/news/8402/nasas-insight-places-first-instrument-on-mars/")
print(f"The Title of the First Article: {first_article}")
```

```python
content_title = soup.find('p').text
print(f"Article Summary: {content_title}")
```
